```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
ds=pd.read_csv('/content/Ads_CTR_Optimisation.csv')
```

```python
ds.head()
```

```python
#total number of ads
d=10
#number of times ad 1 was selected
no_of_selections=[0]*d
#sum of number of times ad 1 was correctly selected
sums_of_reward=[0]*d
#total number of times
N=1000
#the ads algo will select
ads_selected=[]
#total reward
total_reward=0

#for all rounds
for n in range(0,N):
  max_upper_bound=0
  ad=0
  #for all ads
  for i in range(0,d):
    if no_of_selections[i]>0:
      average_reward=sums_of_reward[i]/float(no_of_selections[i])
      delta_i=math.sqrt((3/float(2))*(math.log(n+1)/float(no_of_selections[i])))
      upper_bound=average_reward+delta_i
    else:
      upper_bound=1e400

    if upper_bound>max_upper_bound:
      max_upper_bound=upper_bound
      ad=i
  ads_selected.append(ad)
  no_of_selections[ad]=no_of_selections[ad]+1
  reward=ds.values[n,ad]
  sums_of_reward[ad]+=reward
  total_reward+=reward





```
